```python
gdf = gpd.read_file("vet_walk_polygons.shp")
gdf.head()
```

```python
gdf = gdf.rename(columns = {"Name":"name", "FromBreak": "org_mile", "ToBreak": "dst_mile"})
gdf.head(1)
```

```python
gdf = gdf.drop(["FacilityID"], axis = 1)
gdf.head(1)
```

```python
def rm(x):
    name = x.split(":")[0]
    return name.strip()
```

```python
rm("Verona Veterinary Medical Service : 804 - 1609")
```

```python
gdf["name"] = gdf["name"].apply(lambda x : rm(x))
gdf.head(1)
```

```python
def meter_mile(x):
    if x == float(0):
        return 0
    if x == float(804):
        return 0.5
    if x == float(1609):
        return 1
    else:
        pass
```

```python
gdf["org_mile"] = gdf["org_mile"].apply(lambda x : meter_mile(x))
gdf["dst_mile"] = gdf["dst_mile"].apply(lambda x : meter_mile(x))
gdf.head()
```

```python
os.remove("vet_walk_polygons.shp")
```

```python
gdf.to_file("vet_walk_polygons.shp")
```

```python
ax = gdf.plot(color="#E3DA93", edgecolor="#6E6E6E")
plt.title("1 mile walk polygons from vet clinics in Madison, WI")
```
